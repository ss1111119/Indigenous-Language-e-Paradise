```python
# 歌謠篇_解析網址
import requests
from bs4 import BeautifulSoup

# 定義方言與語言組的對應關係字典
language_groups = {
    1: "南勢阿美語",
    2: "秀姑巒阿美語",
    3: "海岸阿美語",
    4: "馬蘭阿美語",
    5: "恆春阿美語",
    6: "賽考利克泰雅語",
    7: "澤敖利泰雅語",
    8: "汶水泰雅語",
    9: "萬大泰雅語",
    10: "四季泰雅語",
    11: "宜蘭澤敖利泰雅語",
    13: "賽夏語",
    14: "邵語",
    15: "都達賽德克語",
    16: "德固達雅賽德克語",
    17: "德鹿谷賽德克語",
    18: "卓群布農語",
    19: "卡群布農語",
    20: "丹群布農語",
    21: "巒群布農語",
    22: "郡群布農語",
    23: "東排灣語",
    24: "北排灣語",
    25: "中排灣語",
    26: "南排灣語",
    27: "東魯凱語",
    28: "霧台魯凱語",
    29: "大武魯凱語",
    30: "多納魯凱語",
    31: "茂林魯凱語",
    32: "萬山魯凱語",
    33: "太魯閣語",
    34: "噶瑪蘭語",
    35: "鄒語",
    38: "南王卑南語",
    39: "知本卑南語",
    40: "西群卑南語",
    41: "建和卑南語",
    42: "雅美語",
    43: "撒奇萊雅語",
    36: "卡那卡那富語",
    37: "拉阿魯哇語"
}

def get_target_urls():
    target_urls = {}

    for dialect_id in range(1, 44):
        urls_for_dialect = []

        for a in range(1, 11):
            original_url = f"https://web.klokah.tw/extension/song_practice/index.php?d={dialect_id}&s={a}&view=lyric"
            response = requests.get(original_url)

            if response.status_code == 200:
                soup = BeautifulSoup(response.content, 'html.parser')
                iframes = soup.find_all('iframe')

                for iframe in iframes:
                    src = iframe.get('src', '')
                    if 'https://web.klokah.tw/text/read_embed.php?tid=' in src and 'mode=1' in src:
                        urls_for_dialect.append(src)
                        break  # 只需獲取第一個匹配的 URL

                if not urls_for_dialect:
                    urls_for_dialect.append(None)  # 如果未找到匹配的 URL，則將None添加到列表中
            else:
                urls_for_dialect.append(None)  # 如果請求失敗，則將None添加到列表中

        # 將方言ID與對應的目標URL列表以及語言組名稱一起存儲在字典中
        target_urls[dialect_id] = {
            "language_group": language_groups.get(dialect_id, "未知"),  # 獲取語言組名稱，如果不存在則返回"未知"
            "urls": urls_for_dialect
        }

    return target_urls

# 獲取所有方言對應的目標URL
all_target_urls = get_target_urls()

# 打印所有方言對應的目標URL
for dialect_id, data in all_target_urls.items():
    print(f"方言ID {dialect_id}，語言組: {data['language_group']}，的目標URLs:")
    for url in data['urls']:
        print(url)
```
