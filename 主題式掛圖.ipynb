```python
# 主題式掛圖
import requests
import pandas as pd
import re
import json


# 定義方言與語言組的對應關係字典
language_groups = {
    1: "南勢阿美語",
    2: "秀姑巒阿美語",
    3: "海岸阿美語",
    4: "馬蘭阿美語",
    5: "恆春阿美語",
    6: "賽考利克泰雅語",
    7: "澤敖利泰雅語",
    8: "汶水泰雅語",
    9: "萬大泰雅語",
    10: "四季泰雅語",
    11: "宜蘭澤敖利泰雅語",
    13: "賽夏語",
    14: "邵語",
    15: "都達賽德克語",
    16: "德固達雅賽德克語",
    17: "德鹿谷賽德克語",
    18: "卓群布農語",
    19: "卡群布農語",
    20: "丹群布農語",
    21: "巒群布農語",
    22: "郡群布農語",
    23: "東排灣語",
    24: "北排灣語",
    25: "中排灣語",
    26: "南排灣語",
    27: "東魯凱語",
    28: "霧台魯凱語",
    29: "大武魯凱語",
    30: "多納魯凱語",
    31: "茂林魯凱語",
    32: "萬山魯凱語",
    33: "太魯閣語",
    34: "噶瑪蘭語",
    35: "鄒語",
    36: "卡那卡那富語",
    37: "拉阿魯哇語",
    38: "南王卑南語",
    42: "雅美語",
    43: "撒奇萊雅語"
}
# 定義課程名稱列表
course_names = ["身體部位", "親屬稱謂篇", "山川自然景觀篇", "動物篇"]

# 單元類型對應關係
unit_types = {
    "word": "單詞",
    "dial": "對話",
    "game": "遊戲",
    "song": "歌謠"
}

# 用於存儲所有數據的列表

def fetch_and_save_data():
    for b, dialect_name in language_groups.items():
        all_data = []
        for a in range(1, 5):
            url = f"https://web.klokah.tw/flipChart/json/{a}/{b}.json"
            response = requests.get(url)
            if response.status_code == 200:
                # 獲取響應內容作為json_str的值
                json_str = response.text
                # 修正JSON字符串
                json_str = re.sub(r',\s*\]', ']', json_str)
                json_str = re.sub(r',\s*\}', '}', json_str)
                # 解析JSON數據
                try:
                    data = json.loads(json_str)
                    for category_key, items in data.items():
                        category_name = unit_types.get(category_key, "未知")
                        for item in items:
                            entry = {
                                "課名": course_names[a - 1],
                                "方言": dialect_name,
                                "類別": category_name,
                                "中文": item["ch"],
                                "族語": item["ab"]
                            }
                            all_data.append(entry)
                except json.JSONDecodeError as e:
                    print(f"解析JSON失敗: {e}，URL: {url}")

        # 保存數據到Excel
        if all_data:
            df = pd.DataFrame(all_data)
            filename = f"{b}_{dialect_name}.xlsx"
            df.to_excel(filename, index=False)
            print(f"{dialect_name}的數據已保存至 {filename}")
        else:
            print(f"{dialect_name}沒有數據可保存。")

fetch_and_save_data()
```
