```python
# 初級動物、職務、職業，1方言三種類別的excel檔
import requests
import pandas as pd


def fetch_and_save_data():
    language_groups = {
        1: "南勢阿美語",
        2: "秀姑巒阿美語",
        3: "海岸阿美語",
        4: "馬蘭阿美語",
        5: "恆春阿美語",
        6: "賽考利克泰雅語",
        7: "澤敖利泰雅語",
        8: "汶水泰雅語",
        9: "萬大泰雅語",
        10: "四季泰雅語",
        11: "宜蘭澤敖利泰雅語",
        13: "賽夏語",
        14: "邵語",
        15: "都達賽德克語",
        16: "德固達雅賽德克語",
        17: "德鹿谷賽德克語",
        18: "卓群布農語",
        19: "卡群布農語",
        20: "丹群布農語",
        21: "巒群布農語",
        22: "郡群布農語",
        23: "東排灣語",
        24: "北排灣語",
        25: "中排灣語",
        26: "南排灣語",
        27: "東魯凱語",
        28: "霧台魯凱語",
        29: "大武魯凱語",
        30: "多納魯凱語",
        31: "茂林魯凱語",
        32: "萬山魯凱語",
        33: "太魯閣語",
        34: "噶瑪蘭語",
        35: "鄒語",
        38: "南王卑南語",
        39: "知本卑南語",
        40: "西群卑南語",
        41: "建和卑南語",
        42: "雅美語",
        43: "撒奇萊雅語",
        36: "卡那卡那富語",
        37: "拉阿魯哇語"
}

    for dialect_id, dialect_name in language_groups.items():
        # 定義需要處理的類別及其對應的URL結構
        categories = {
            '動物': f"https://web.klokah.tw/mode/json/{dialect_id:02d}/elementary/animal.json",
            '植物': f"https://web.klokah.tw/mode/json/{dialect_id:02d}/elementary/plant.json",
            '職業': f"https://web.klokah.tw/mode/json/{dialect_id:02d}/elementary/feature.json"
        }

        # 打開一個ExcelWriter來寫入數據
        filename = f"{dialect_id}_{dialect_name}_初級.xlsx"
        with pd.ExcelWriter(filename) as writer:
            for category, url in categories.items():
                response = requests.get(url)
                if response.status_code == 200:
                    json_data = response.json()
                    theme = json_data.get("theme", "未知主題")
                    all_data = []

                    # 提取不同階段的數據
                    stages = ['opening', 'phase1', 'phase2', 'phase3',
                              'phase4', 'phase5', 'phase6', 'switch', 'ending']
                    for stage in stages:
                        stage_data = json_data['text'].get(stage, {})
                        if stage_data:
                            ab = stage_data.get('ab', '')
                            ch = stage_data.get('ch', '')
                            all_data.append({'階段': stage, '族語': ab, '中文': ch})

                    # 提取Phase 2的數據
                    phase2_items = json_data['data'].get('phase2', [])
                    for item in phase2_items:
                        phase_ab = item.get('text', '')
                        phase_ch = item.get('textCh', '')
                        all_data.append(
                            {'階段': item.get('name', '未知類型'), '族語': phase_ab, '中文': phase_ch})

                    # 數據處理完成後，轉換為DataFrame並存儲於適當的工作表
                    if all_data:
                        df = pd.DataFrame(all_data)
                        df.to_excel(writer, sheet_name=category, index=False)
                        print(
                            f"{dialect_name}的{category}數據已保存至 {filename} 的 {category} 工作表中")
                    else:
                        print(f"沒有{category}數據可保存。")
                else:
                    print(f"請求{category}失敗，狀態碼: {response.status_code}")


# 執行函數
fetch_and_save_data()
```

```python
import requests
import pandas as pd

def fetch_and_save_new_data():
    language_groups = {
        1: "南勢阿美語",
        2: "秀姑巒阿美語",
        3: "海岸阿美語",
        4: "馬蘭阿美語",
        5: "恆春阿美語",
        6: "賽考利克泰雅語",
        7: "澤敖利泰雅語",
        8: "汶水泰雅語",
        9: "萬大泰雅語",
        10: "四季泰雅語",
        11: "宜蘭澤敖利泰雅語",
        13: "賽夏語",
        14: "邵語",
        15: "都達賽德克語",
        16: "德固達雅賽德克語",
        17: "德鹿谷賽德克語",
        18: "卓群布農語",
        19: "卡群布農語",
        20: "丹群布農語",
        21: "巒群布農語",
        22: "郡群布農語",
        23: "東排灣語",
        24: "北排灣語",
        25: "中排灣語",
        26: "南排灣語",
        27: "東魯凱語",
        28: "霧台魯凱語",
        29: "大武魯凱語",
        30: "多納魯凱語",
        31: "茂林魯凱語",
        32: "萬山魯凱語",
        33: "太魯閣語",
        34: "噶瑪蘭語",
        35: "鄒語",
        38: "南王卑南語",
        39: "知本卑南語",
        40: "西群卑南語",
        41: "建和卑南語",
        42: "雅美語",
        43: "撒奇萊雅語",
        36: "卡那卡那富語",
        37: "拉阿魯哇語"
    }

    topics = {
        '主題一邀約': [
            "https://web.klokah.tw/kongmode/JSON/{dialect_id}/A/main.json",
            "https://web.klokah.tw/kongmode/JSON/{dialect_id}/A/audio.json",
            "https://web.klokah.tw/kongmode/JSON/{dialect_id}/instructions.json"
        ],
        '主題二-訪談用語': [
            "https://web.klokah.tw/kongmode/JSON/{dialect_id}/B/main.json",
            "https://web.klokah.tw/kongmode/JSON/{dialect_id}/B/audio.json",
            "https://web.klokah.tw/kongmode/JSON/{dialect_id}/instructions.json"
        ],
        '主題三-農耕活動': [
            "https://web.klokah.tw/kongmode/JSON/{dialect_id}/C/main.json",
            "https://web.klokah.tw/kongmode/JSON/{dialect_id}/C/audio.json",
            "https://web.klokah.tw/kongmode/JSON/{dialect_id}/instructions.json"
        ]
    }

    for dialect_id, dialect_name in language_groups.items():
        filename = f"{dialect_id}_{dialect_name}_中高級.xlsx"
        with pd.ExcelWriter(filename) as writer:
            for topic, urls in topics.items():
                all_data = []
                formatted_urls = [url.format(dialect_id=dialect_id) for url in urls]
                for url in formatted_urls:
                    try:
                        response = requests.get(url, timeout=10)  # 添加超時
                        response.raise_for_status()  # 檢查響應狀態碼
                        json_data = response.json()
                        for item in json_data:
                            data_entry = {'n': item.get('n'), 'ab': item.get('ab'), 'ch': item.get('ch')}
                            if 'keyw' in item:
                                data_entry.update({'n': item.get('n'),'keyw': item.get('keyw'), 'keych': item.get('keych')})
                            all_data.append(data_entry)
                    except requests.RequestException as e:
                        print(f"請求 {url} 失敗: {e}")

                if all_data:
                    df = pd.DataFrame(all_data)
                    df.to_excel(writer, sheet_name=topic[:31], index=False)  # 確保工作表名不超過31個字符
                    print(f"{dialect_name}的{topic}數據已保存至 '{filename}' 的 '{topic}' 工作表中")
                else:
                    print(f"沒有{topic}數據可保存。")

# 執行函數
fetch_and_save_new_data()
```
