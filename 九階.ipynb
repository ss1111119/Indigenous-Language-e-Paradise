```python
import requests
import pandas as pd

# 定義方言及其對應的編號
dialects = {
    "24": "南勢阿美",
    "23": "秀姑巒阿美",
    "25": "海岸阿美",
    "26": "馬蘭阿美",
    "27": "恆春阿美",
    "16": "賽考利克泰雅",
    "17": "澤敖利泰雅",
    "18": "汶水泰雅語",
    "19": "萬大泰雅語",
    "43": "四季泰雅語",
    "42": "宜蘭澤敖利泰雅",
    "28": "賽夏語",
    "30": "邵語",
    "22": "都達賽德克語",
    "20": "德固達雅賽德克語",
    "21": "德鹿谷賽德克語",
    "6": "卓群布農語",
    "7": "卡群布農語",
    "8": "丹群布農語",
    "9": "巒群布農語",
    "5": "郡群布農語",
    "11": "東排灣語",
    "12": "北排灣語",
    "13": "中排灣語",
    "10": "南排灣語",
    "15": "東魯凱語",
    "14": "霧台魯凱語",
    "41": "大武魯凱語",
    "35": "多納魯凱語",
    "37": "多納魯凱語",
    "36": "多納魯凱語",
    "44": "太魯閣語",
    "31": "噶瑪蘭語",
    "32": "鄒語",
    "33": "卡那卡那富語",
    "34": "拉阿魯哇語",
    "3": "南王卑南語",
    "2": "知本卑南語",
    "1": "西群卑南語",
    "4": "建和卑南語",
    "29": "雅美語",
    "38": "撒奇萊雅語"
}

# 方言總數，用於顯示進度
total_dialects = len(dialects)
current_dialect = 1

# 迴圈遍歷方言編號和名稱
for d, name in dialects.items():
    print(f"開始處理 {name} ({current_dialect}/{total_dialects})，方言編號：{d}")

    # 準備數據列表
    data_list = []

    # 迴圈l=1到l=9，對於每個l值
    for l in range(1, 10):
        # 迴圈c=1到c=10，對於每個c值
        for c in range(1, 11):
            print(f"處理中... l={l}, c={c}")

            # 構造請求的URL
            url = f"https://web.klokah.tw/ninew/php/getTextNew.php?d={d}&l={l}&c={c}"

            # 發起GET請求
            response = requests.get(url)

            if response.status_code == 200:
                # 解析返回的JSON數據
                data = response.json()

                # 遍歷句子並收集數據
                for sentence in data["sentence"]:
                    row = {
                        "C": c,  # 標識當前c的值
                        "L": l,  # 標識當前l的值
                        "Order": sentence["order"]
                    }

                    # 為每個單詞添加列
                    for i, word in enumerate(sentence["word"], start=1):
                        row[f"Word_{i}"] = word["ab"]

                    data_list.append(row)
            else:
                print(f"請求d={d}, c={c}, l={l}失敗，狀態碼:", response.status_code)

    # 轉換為DataFrame
    df = pd.DataFrame(data_list)

    # 構建文件名
    filename = f"{d}_{name}.xlsx"
    # 將DataFrame保存為Excel文件
    df.to_excel(filename, index=False)

    print(f"{d}_{name}的數據已保存至 {filename}")
    current_dialect += 1
```
