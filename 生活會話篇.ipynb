```python
# 生活會話篇
import requests
import xml.etree.ElementTree as ET
import pandas as pd

# 定義語言編號和名稱
languages = [
    (1, "南勢阿美語"),
    (2, "秀姑巒阿美語"),
    (3, "海岸阿美語"),
    (4, "馬蘭阿美語"),
    (5, "恆春阿美語"),
    (6, "賽考利克泰雅語"),
    (7, "澤敖利泰雅語"),
    (8, "汶水泰雅語"),
    (9, "萬大泰雅語"),
    (10, "四季泰雅語"),
    (11, "宜蘭澤敖利泰雅語"),
    (13, "賽夏語"),
    (14, "邵語"),
    (15, "都達賽德克語"),
    (16, "德固達雅賽德克語"),
    (17, "德鹿谷賽德克語"),
    (18, "卓群布農語"),
    (19, "卡群布農語"),
    (20, "丹群布農語"),
    (21, "巒群布農語"),
    (22, "郡群布農語"),
    (23, "東排灣語"),
    (24, "北排灣語"),
    (25, "中排灣語"),
    (26, "南排灣語"),
    (27, "東魯凱語"),
    (28, "霧台魯凱語"),
    (29, "大武魯凱語"),
    (30, "多納魯凱語"),
    (31, "茂林魯凱語"),
    (32, "萬山魯凱語"),
    (33, "太魯閣語"),
    (34, "噶瑪蘭語"),
    (35, "鄒語"),
    (38, "南王卑南語"),
    (39, "知本卑南語"),
    (40, "西群卑南語"),
    (41, "建和卑南語"),
    (42, "雅美語"),
    (43, "撒奇萊雅語"),
    (36, "卡那卡那富語"),
    (37, "拉阿魯哇語")
]

# 遍歷每種語言
for language_id, language_name in languages:
    url = f"https://web.klokah.tw/extension/con_data/xml/{language_id}/conversation.xml"

    response = requests.post(url)

    if response.status_code == 200:
        root = ET.fromstring(response.content)

        data_list = []  # 初始化數據列表

        # 處理課程信息
        lessons = root.findall('lesson')
        for lesson in lessons:
            lessonId = lesson.find('lessonId').text
            lessonAB = lesson.find('lessonAB').text
            lessonCH = lesson.find('lessonCH').text

            # 處理對話內容
            contents = root.findall(f".//content[lessonId='{lessonId}']")
            for content in contents:
                for i in range(1, 13):  # 假設最多12個場景
                    scene_tag = f'con_scene{i}'
                    ab_tag = f'con_AB{i}'
                    ch_tag = f'con_CH{i}'

                    scene = content.find(scene_tag)
                    ab = content.find(ab_tag)
                    ch = content.find(ch_tag)

                    if scene is not None and ab is not None and ch is not None:
                        data_list.append({
                            "課程ID": lessonId,
                            "阿美語課程": lessonAB,
                            "中文課程": lessonCH,
                            "場景": scene.text,
                            "阿美語對話": ab.text,
                            "中文對話": ch.text
                        })

        # 轉換為DataFrame
        df = pd.DataFrame(data_list)

        # 構建文件名並保存為Excel文件
        filename = f"{language_id}_{language_name}.xlsx"
        df.to_excel(filename, index=False)
        print(f"{language_name}的數據已保存至 {filename}")
    else:
        print(f"請求語言{language_name}失敗，狀態碼: {response.status_code}")
```
